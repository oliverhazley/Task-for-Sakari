<h2 style="text-align: center;"> Havainnot </h2>

**Syke ja RMSSD (parasympaattinen aktiivisuus):**
- Syke vaihtelee välillä 62-75 bpm, ja keskiarvo on 69.9.
- RMSSD (parasympaattisen hermoston aktiivisuuden mittari) korkein arvo on 46 ms, ja laskee myöhemmissä mittauksissa 28. Keskiarvo on 37.2 ms.

**Stressi-indeksi ja LF/HF-suhde:**
- Stressi-indeksi vaihtelee mittauksissa, ja korkeimmat arvot (yli 10) viittaavat kohonneeseen fysiologiseen stressiin.
- LF/HF-suhde, joka mittaa sympaattisen ja parasympaattisen hermoston tasapainoa, nousee joissakin mittauksissa merkittävästi (6.36, 9.15), mikä viittaa sympaattisen hermoston (stressivasteen) hallitsevuuteen.

**Autonominen tasapaino (SNS vs. PNS -indeksit):**
- SNS-indeksi (sympaattinen hermosto) muuttuu negatiivisista arvoista positiivisiin, mikä osoittaa stressivasteen voimistumista.
- PNS-indeksi (parasympaattinen hermosto) alkaa positiivisena mutta muuttuu negatiiviseksi, mikä kertoo palautumisen heikkenemisestä ja mahdollisesti kertyneestä stressistä.

**Tulkinnat:**
- Alkukantaan verrattuna keho on myöhemmissä mittauksissa vähemmän palautunut, ja stressitaso on kasvanut.
- RMSSD:n lasku ja LF/HF-suhteen nousu viittaavat siihen, että palautumiskyky on heikentynyt mittausjakson aikana.
- Kasvavat stressi-indeksit ja laskeva parasympaattinen aktiivisuus voivat kertoa siitä, että keho ei ole saanut riittävästi lepoa.

Kubiosin HRV-sovelluksessa valmius (readiness) ja fysiologinen ikä (physiological age) ovat tärkeitä mittareita, jotka auttavat arvioimaan kehon palautumista, stressitasoa ja yleistä hyvinvointia.

<br>
<br>
<br>
<h2 style="text-align: center;"> Valmius (Readiness) </h2>
<br>

**Määritelmä:**
- Valmiusindeksi kuvastaa kehon fysiologista palautumista ja stressitasoa. Se perustuu sykevälivaihtelun (HRV) mittauksiin, erityisesti RMSSD-arvoon, joka heijastaa parasympaattisen hermoston aktiivisuutta.

**Tulkinta:**
- Korkea valmiusindeksi: Viittaa hyvään palautumiseen, matalaan stressitasoon ja korkeaan valmiuteen fyysisiin tai henkisiin haasteisiin.
- Matala valmiusindeksi: Voi osoittaa riittämätöntä palautumista, korkeampaa stressitasoa tai ylikuormitusta, mikä saattaa edellyttää kevyempää harjoittelua tai lisääntynyttä lepoa.

**Vaikuttavat tekijät:**
- **Uni:** Laadukas ja riittävä uni parantaa palautumista ja nostaa valmiusindeksiä.
- **Stressi:** Korkea stressitaso voi laskea valmiusindeksiä heikentämällä palautumista.
- **Fyysinen aktiivisuus:** Sopiva harjoittelu parantaa valmiutta, mutta liiallinen kuormitus ilman riittävää palautumista voi laskea indeksiä.
- **Ravitsemus ja nesteytys:** Tasapainoinen ruokavalio ja riittävä nesteytys tukevat optimaalista palautumista.

<br>
<br>
<br>
<h2 style="text-align: center;"> Fysiologinen ikä (Physiological Age) </h2>
<br>

**Määritelmä:**
- Fysiologinen ikä arvioi kehon toiminnallista tilaa suhteessa kronologiseen ikään. Se perustuu leposykevälivaihtelun analyysiin, erityisesti RMSSD-arvoon, ja vertaa sitä väestön normaaliarvoihin.

**Tulkinta:**
- Fysiologinen ikä **<** kronologinen ikä: Kehon toiminnallinen tila on parempi kuin keskimäärin kyseisessä ikäryhmässä.
- Fysiologinen ikä **≈** kronologinen ikä: Kehon toiminnallinen tila vastaa ikäryhmän keskiarvoa.
- Fysiologinen ikä **>** kronologinen ikä: Kehon toiminnallinen tila on heikompi kuin keskimäärin kyseisessä ikäryhmässä, mikä saattaa viitata tarpeeseen parantaa elämäntapoja.

**Vaikuttavat tekijät:**
- Elämäntavat (säännöllinen liikunta, terveellinen ravinto, riittävä uni)
- Stressinhallinta (esim. mindfulness, jooga)
- Sairaudet ja terveydentila

<br>

<h2 style="text-align: center;"> Autonomisen hermoston toiminnan analyysi </h2>
<br>

**Testit:**
- **Valsalvan koe:** Potilas puhaltaa voimakkaasti suljettuun putkeen, mikä aiheuttaa muutoksia sydämen sykkeessä ja verenpaineessa. Tämä testi arvioi autonomisen hermoston refleksivastetta.
- **Syvähengitystesti:** Potilas hengittää syvään ja hitaasti, ja sykevaihtelua mitataan hengityssyklin aikana. Tämä testi arvioi parasympaattisen hermoston toimintaa.
- **Pystyynnousutesti (Head-Up Tilt):** Potilas siirretään makuuasennosta pystyasentoon, ja sydämen syke ja verenpaine mitataan. Tämä testi arvioi autonomisen hermoston kykyä säädellä verenkiertoa asennon muutoksissa.

**Vaikutus HRV-parametreihin:**
- Valsalvan koe: Syke nousee ensin ja laskee sitten, mikä heijastaa tasapainoista autonomista toimintaa.
- Syvähengitystesti: Suurempi sykevaihtelu hengityksen aikana viittaa terveeseen parasympaattiseen toimintaan.
- Pystyynnousutesti: Nopea sykenousu ja sen jälkeinen palautuminen normaaliin viittaavat terveeseen autonomiseen säätelyyn.

### 2. Python-toteutus (OAuth2 Client Credentials)
Alla on Python-esimerkki, joka hakee access_tokenin ja käyttää sitä API-kutsussa:


    import requests

    # KubiosCloud autentikointitiedot
    auth_url = "https://kubioscloud.auth.eu-west-1.amazoncognito.com/oauth2/token"
    client_id = "your_client_id"
    client_secret = "your_client_secret"
    
    # Haetaan OAuth2 access_token
    data = {"grant_type": "client_credentials"}
    response = requests.post(auth_url, data=data, auth=(client_id, client_secret))
    
    if response.status_code == 200:
        token = response.json().get("access_token")
        print("Access Token:", token)
    else:
        print("Authentication Failed:", response.text)
    
    # API-kutsu käyttäjän tietojen hakemiseksi
    headers = {"Authorization": f"Bearer {token}"}
    api_url = "https://analysis.kubioscloud.com/v1/user/self"
    api_response = requests.get(api_url, headers=headers)
    
    print(api_response.json())




**Miten tämä toimii?**
1. POST-pyyntö lähetetään autentikointipalvelimelle käyttäjätiedoilla.
2. Palvelin palauttaa access_tokenin, jota käytetään kaikissa API-kutsuissa.

<h2 style="text-align: center;"> Päätoiminnot </h2>

**Käyttäjän autentikaatio**
- Käyttäjä kirjautuu sisään sähköpostilla ja salasanalla.
- Backend validoi käyttäjän ja generoi JWT-tokenin, joka tallennetaan sessioon.

**HRV-seuranta ja laiteyhteys**
- Sovellus yhdistyy Polar H10 tai Oura Ring -laitteeseen Bluetoothin avulla.
- Mitattu HRV-data lähetetään backendille ja tallennetaan MariaDB-tietokantaan.

**Pelillistetty terapia ja harjoitukset**
- Käyttäjä voi pelata Tetris-peliä, joka tukee stressinhallintaa.
- Mindfulness- ja hengitysharjoitukset tarjoavat rauhoittumismenetelmiä.
- Käyttäjän suorittamat harjoitukset ja pelitulokset tallennetaan MariaDB-tietokantaan.

**Asetukset ja laitehallinta**
- Käyttäjä voi yhdistää ja hallita HRV-laitteita.
- Asetusten kautta käyttäjä voi määrittää henkilökohtaiset terapiatavoitteet.

**Datan käsittely ja analyysi**
- Käyttäjä voi tarkastella HRV-mittauksia ja stressitasoja graafisesti.
- Data voidaan jakaa terapeutin kanssa analysointia varten.

MariaDB valittiin tietokannaksi, koska se tarjoaa tehokkaan relaatiotietokantaratkaisun, tukee monimutkaisia kyselyjä ja skaalautuu hyvin sovelluksen tarpeisiin. Tämä malli mahdollistaa reaaliaikaisen HRV-datan seurannan, terapeuttisten harjoitusten käytön ja käyttäjätietojen tallennuksen turvallisesti.
